```python
import pandas as pd
import numpy as np
import math

UNCLASSIFIED = False
NOISE = -1
```

euclidean distance formula

```python
def _dist(p, q):
    # compute and return the euclidean distance
    # put your code here
    return math.sqrt(math.sqrt(np.power(p-q,2).sum()))
```

`eps`: the distance to decide if another point is neighbor (their circle)

```python
def _eps_neighborhood(p, q, eps: float):
    # check if the distance between p and q is below eps
    # return True or False
    # put your code here
    return _dist(p,q) < eps
```

```python
# for better typing's hint
from typing import TypeVar
Numpy_Matrix = TypeVar('numpy.matrix')
```

```python
def _region_query(m: Numpy_Matrix, point_id, eps: float) -> list:
    n_points = m.shape[1]
    seeds = []
    # find and return all points that belong to eps-neighborhood of point_id
    # put your code here
    for i in range(0, n_points):
        if _eps_neighborhood(m[:, point_id], m[:, i], eps):
            seeds.append(i)
    return seeds
```

`min_points`: the number of neighbors needed to say a region is dense

`core`: a point with at least MinPts neighbors (very high-dense point)

`Noise`: the set of points in database **NOT** belonging to any cluster

###### steps of dbscan:

1. Start with an arbitrary point p from data, and retrieve all points density‐reachable from p
- w.r.t. Eps and MinPts

2. If p is a core point, generate a cluster 
- w.r.t. Eps and MinPts, and all points in the cluster are labeled

3. If p is a border point, no points are density‐reachable from p

***Repeat (1) (2) (3) by visiting the next unlabeled point in data***

```python
def _expand_cluster(m: Numpy_Matrix, classifications, point_id, cluster_id, eps: float, min_points: int) -> bool:
    # expand the cluster of cluster_id from point_id
    # identify all points belonging to cluster_id
    # update the clustering results in "classification" by assigning cluster_id to each point
    # return True if point_id is a core, False if point_id is not a core
    # write your code below
    seeds = _region_query(m, point_id, eps)
    # mark the point as a noise, and return False if the length of seeds list is less than min_points
    # note that `UNCLASSIFIED` = False & `NOISE` = -1
    if len(seeds) < min_points:
        classifications[point_id] = NOISE
        return False
    else:
        classifications[point_id] = cluster_id
        for id_of_seed in seeds:
            classifications[id_of_seed] = cluster_id
        
        # updating the results untill the expand is converge
        while len(seeds) > 0:
            point = seeds[0]
            results = _region_query(m, point, eps)
            if len(results) >= min_points:
                for i in range(0, len(results)):
                    result_point = results[i]
                    if classifications[result_point] == UNCLASSIFIED: seeds.append(result_point)
                    classifications[result_point] = cluster_id
            seeds = seeds[1:]
        return True
```

[wiki dbscan](https://en.wikipedia.org/wiki/DBSCAN)

```python
def dbscan(m: Numpy_Matrix, eps: float, min_points: int) -> list:
    """Implementation of DBSCAN
    You can refer to wikipedia for detailed algorithm: https://en.wikipedia.org/wiki/DBSCAN
    Use Euclidean Distance as the measure
    
    Inputs:
    m - A matrix whose columns are feature vectors
    eps - Maximum distance two points can be to be regionally related
    min_points - The minimum number of points to make a cluster
    
    Outputs:
    An array with either a cluster id number or dbscan.NOISE (None) for each column vector in m
    """
    # note that `UNCLASSIFIED` = False & `NOISE` = -1
    cluster_id = 1
    n_points = m.shape[1]
    classifications = [UNCLASSIFIED] * n_points
    # the main dbscan algorithm
    # put your code here
    for point_id in range(n_points):
        point = m[:, point_id]
        if classifications[point_id] == UNCLASSIFIED and _expand_cluster(m, classifications, point_id, cluster_id, eps, min_points):
            cluster_id = cluster_id + 1
    from collections import Counter
    print(f'classification: {Counter(classifications)}')
    return classifications
```
