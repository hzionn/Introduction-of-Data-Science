```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cdata.csv')
xs = [i for i in df.x]
ys = [i for i in df.y]
input_data = [[i, j] for i, j in zip(xs, ys)]
labels = [i for i in df.cluster]
input_data
```

```python
input_data = np.array([[i, j] for i, j in zip(xs, ys)])
input_data
```

```python
cluster = KMeans(n_clusters=4, random_state=0)
model = cluster.fit(input_data)
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
plt.scatter(xs, ys)
plt.scatter([i[0] for i in model.cluster_centers_], [i[1] for i in model.cluster_centers_], marker='*', s=200, color='gold')
plt.show()
```
