```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, accuracy_score, f1_score, balanced_accuracy_score, plot_confusion_matrix
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
import os
```

```python
PATH = '/Users/zionn/Desktop/資料科學導論/hw5/'
os.getcwd()
```

```python
PATH_TRAIN = PATH + 'churn_prediction_data/train.csv'
df_train = pd.read_csv(PATH_TRAIN)
#df_train.head()
```

```python
cols = ['CreditScore', 'Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary', 'Exited']
data_dummy = pd.get_dummies(df_train.loc[:, cols])
df_train = pd.DataFrame(data_dummy)
df_train['HasCrCard'] = df_train['HasCrCard'].replace({0: -1})
df_train['IsActiveMember'] = df_train['IsActiveMember'].replace({0: -1})
#df_train.head()
```

```python
train_label = df_train.loc[:, 'Exited']
df_train.drop('Exited', axis=1, inplace=True)
train_data = df_train.loc[:, 'CreditScore': 'Gender_Male']
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
ros = RandomOverSampler()
smote = SMOTE()
train_data, train_label = smote.fit_resample(train_data, train_label)
```

```python
scaler = StandardScaler()
x_train_standardized = scaler.fit_transform(train_data)
y_train = train_label
```

```python
#x_train_standardized, x_test_standardized, y_train, y_test = train_test_split(x_train_standardized, y_train, train_size = 6000, test_size = 2000, random_state = 48)
```

```python
clf = RandomForestClassifier(n_jobs = -1)
clf.fit(x_train_standardized, y_train)
```

```python
check_train = clf.predict(x_train_standardized)
accuracy_score(y_train, check_train), precision_score(y_train, check_train), f1_score(y_train, check_train)
```

```python
PATH_TEST = PATH + 'churn_prediction_data/test.csv'
df_test = pd.read_csv(PATH_TEST)
rw = [i for i in df_test.RowNumber]
```

```python
data_dummy = pd.get_dummies(df_test.loc[:, 'CreditScore': 'EstimatedSalary'])
df_test = pd.DataFrame(data_dummy)
df_test['HasCrCard'] = df_test['HasCrCard'].replace({0: -1})
df_test['IsActiveMember'] = df_test['IsActiveMember'].replace({0: -1})
```

```python
x_hat_test = df_test.loc[:, 'CreditScore': 'Gender_Male']
x_hat_test_standardized = scaler.transform(x_hat_test)
```

```python
y_hat_predictions_label = clf.predict(x_hat_test_standardized)
y_hat_predictions_label
```

```python
df_upload = pd.DataFrame({
    'RowNumber': rw, 
    'Exited': y_hat_predictions_label
})

df_upload.to_csv(PATH + 'churn_prediction_data/RF_8000.csv')
```

```python
PATH = '/Users/zionn/Desktop/資料科學導論/hw5/'
df_dnn = pd.read_csv(PATH + 'churn_prediction_data/DNN_8000.csv')
df_svm = pd.read_csv(PATH + 'churn_prediction_data/SVM_8000.csv')
df_rf = pd.read_csv(PATH + 'churn_prediction_data/RF_8000.csv')
df_cat = pd.read_csv(PATH + 'churn_prediction_data/CatBoost.csv')
df_xg = pd.read_csv(PATH + 'churn_prediction_data/XGBoost.csv')
df_rfp = pd.read_csv(PATH + 'churn_prediction_data/RandomForest_pan.csv')

import random
count = 0
vote = []
for dnn, svm, rf, cat, xg, rfp in zip(df_dnn.Exited, df_svm.Exited, df_rf.Exited, df_cat.Exited, df_xg.Exited, df_rfp.Exited):
    if dnn == svm == rf == cat == xg == rfp:
        count += 1
    if dnn + svm + rf + cat + xg + rfp > 3:
        vote.append(1)
    elif dnn + svm + rf + cat + xg + rfp < 3:
        vote.append(0)
    elif dnn + svm + rf + cat + xg + rfp == 3:
        vote.append(random.choice([0, 1]))
    
print(count)
```

```python
df_upload = pd.DataFrame({
    'RowNumber': rw, 
    'Exited': vote
})

df_upload.to_csv(PATH + 'churn_prediction_data/vote_8000.csv')
```
